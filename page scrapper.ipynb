```python
soup = BeautifulSoup(r.text, "html.parser")
```

```python
soup
```

```python
len(soup.find_all("span", class_= "text"))
```

```python
quote = []
for i in soup.find_all("span", class_= "text"):
    quote.append(i.text[1:-1])
quote
```

```python
author_name  =[]
for name in soup.find_all("small",class_ = "author"):
    author_name.append((name.text))
# author_name =",".join(author_name)
author_name
```

```python
author_id  =[]
for i in soup.find_all("div",class_ = "quote"):
    author_id.append(i.find("a").get("href"))
author_id
```

```python
for i in soup.find_all("div",class_= "quote"):
    for j in soup.find_all("a",class_ = "tag"):
        print(j.text)
```
